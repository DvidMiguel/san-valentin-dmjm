EJERCICIO:
Crea un script que:
1. Cree una lista de números del 1 al 10.
2. Cree una función que calcule el cuadrado.
3. Recorra la lista.
4. Si el número es par:
    Guardarlo en un diccionario.
    El valor será su cuadrado.
5. Mostrar el resultado final.

mi solucion

```python
lista = []
for x in range(1,11):
    #print(x)
    lista.append(x)
#print(lista)

def genera_lista_cuadrados(lista_convertir: list):
    diccionario = {}
    for i in lista_convertir:
        if i % 2 == 0:
            #print(f"{i} es par")
            diccionario.update({i: i*i})
        
    #print(lista_cuadrados)
    return diccionario

fin_dict = genera_lista_cuadrados(lista)
print(fin_dict)
```

version revisada y modificada

```python
lista = list(range(1, 11))

def generar_diccionario_cuadrados(lista_convertir: list) -> dict:
    resultado = {i: i**2 for i in lista_convertir if i % 2 == 0}
    return resultado

fin_dict = generar_diccionario_cuadrados(lista)
print(fin_dict)
```
